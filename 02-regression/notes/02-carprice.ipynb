# Linear regression

Model used for predicting numbers

```python
df_train.iloc[10]
```

Example of g(x_i)=y_i where y_i is the predicted price of a car, and x_i are its features (array)
We take only hp, mpg and popularity to simplify

i=10


```python
x_i=[453, 11, 86]
```

```python
# We're looking for a function g that produces y_i from x_i
w0 = 7.17 #bias term
w=[0.01, 0.04, 0.002]
def g(x_i):
    n = len(x_i)
    pred = w0
    for j in range(n):
        pred += w[j] * x_i[j]
    return pred
```

Generally the approach is
g(x_i) = W0 + W1*x_i1 + W2*x_i2 + W3*x_i3 + ...

Each feature gets its own weight

```python
g(x_i)
```

```python
# We undo the logarithm to get the predicted price
np.expm1(g(x_i))
```

## Generalizing the linear regression model

Observation: after the bias weight, the sum of the feature terms that follows is nothing but a dot product

g(x_i) = w_0 + SUM(j=1..n, x_ij*w_j)

    = w_0+x_iTw

```python
def linear_regression(x_i):
    return w0 + np.dot(x_i, w)
```

The bias weight is just hanging around outside. Can we bring it in? Yes. We add a "fake" feature zero set to 1, and we do w_0*1
So:

    w=[w_0, w_1,...w_n]
    x_i=[1, x_i1,...x_in]

```python
w_new = [w0] + w     # include bias weight in weights
def linear_regression(x_i):
    x_i_new = [1] + x_i  # add the fake feature for bias
    return np.dot(x_i_new, w_new)
```

```python
w_new
```

```python
linear_regression(x_i)
```

Now we generalize to all rows of the dataset. Each row must be dot-multiplied with its weights vector.
But this is just a matrix dot product.

But how do we come up with those weights?

# Training a linear regression model

Our situation is g(X) = Xw, and we want Xwy

Basically we need to solve the equation system for w.

If we assume X is invertible, w = inv(X)y

But X is usually not square, so we can't just invert it. The solution doesn't exist, but we can look for an approximate solution:

XTXw=XTy, where XTX is the gram matrix, and this XTX product is square, so inv(XTX) might exist.

Therefore w = (XTX)^-1 XT y

Which allows us to find the w vector.

```python
# example
X_features = [[148,24,1385],
     [132,25,2031],
     [453,11,86],
     [158,24,185],
     [172,25,201],
     [413,11,86],
     [38,54,185],
     [142,25,431],
     [453,31,86]]
X_features = np.array(X_features)
X_features
```

```python
# add the 1-column for bias term
ones = np.ones(X_features.shape[0])
X = np.column_stack([ones, X_features])
```

```python
XTX = X.T.dot(X) # gram matrix
XTX
```

```python
y = [10000, 20000, 15000, 20050, 10000, 20000, 15000, 25000, 12000]
```

```python
XTX_inv = np.linalg.inv(XTX)
XTX_inv
```

```python
# It's actually the inverse of the gram matrix
XTX.dot(XTX_inv).round(1)
```

```python
w_full = XTX_inv.dot(X.T).dot(y)
w_full
```

```python
w0 = w_full[0]
w = w_full[1:]
w0, w
```

```python
# Putting it all together in a function
def train_linear_regression(X, y):
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])

    XTX = X.T.dot(X) # gram matrix
    XTX_inv = np.linalg.inv(XTX)
    w_full = XTX_inv.dot(X.T).dot(y)

    return w_full[0], w_full[1:]  # return bias and weights separately
```

```python
train_linear_regression(X_features, y)
```

= root mean square error.

```python
def rmse(y, y_pred):
    se = (y - y_pred) ** 2
    mse = se.mean()
    return np.sqrt(mse)
```

# Regularization

w = (XTX)^-1  XT  y

Problem: sometimes the inverse doesn't exist, because at least one pair of columns isn't linearly independent.

But if we add a tiny bit of noise to the diagonal, the matrix becomes invertible. This is called regularization, we add very small values to the diagonal to control. 

```python
# Putting it all together in a function
def train_linear_regression_reg(X, y, r=0.001):
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])

    XTX = X.T.dot(X) # gram matrix
    
    XTX = XTX + r * np.eye(XTX.shape[0])  # regularization step

    XTX_inv = np.linalg.inv(XTX)
    w_full = XTX_inv.dot(X.T).dot(y)

    return w_full[0], w_full[1:]  # return bias and weights separately
```

```python
# let's retrain with regularization
X_train = prepare_X(df_train)
w0, w = train_linear_regression_reg(X_train, y_train, r=0.01)

# again, we validate on the validation dataset
X_val = prepare_X(df_val)
y_pred = w0 + X_val.dot(w)

rmse(y_val, y_pred)
```

Considerable improvement!

```python
sts.histplot(y_pred, color='red', bins=50, alpha=0.5)
sts.histplot(y_val, color='blue', bins=50, alpha=0.5)
```

# Tuning the model

The r parameter affects the quality of my model - how do I find the best value?

```python
for r in [0.0, 0.00001, 0.0001, 0.001, 0.01, 0.1, 1, 10]:
    X_train = prepare_X(df_train)
    w0, w = train_linear_regression_reg(X_train, y_train, r=r)

    # again, we validate on the validation dataset
    X_val = prepare_X(df_val)
    y_pred = w0 + X_val.dot(w)

    score = rmse(y_val, y_pred)
    print('r=%.5f %.4f %.8f' % (r, w0, score))
```

0.001 looks good...

```python
r = 0.001

X_train = prepare_X(df_train)
w0, w = train_linear_regression_reg(X_train, y_train, r=r)

X_val = prepare_X(df_val)
y_pred = w0 + X_val.dot(w)

rmse(y_val, y_pred)
```

Now we should test this against the test dataset...

# Using the model

Now that we have a good candidate, we take the train and val datasets together and do a full train on this enlarged 60+20 dataset, and test against the remaining 20% of the test dataset.

```python
df_full_train = pd.concat([df_train, df_val])
df_full_train.reset_index(drop=True, inplace=True)

X_full_train = prepare_X(df_full_train)
y_full_train = np.concatenate([y_train, y_val])

w0, w = train_linear_regression_reg(X_full_train, y_full_train, r=0.001)
```

```python
X_test = prepare_X(df_test)
y_pred = w0 + X_test.dot(w)
score = rmse(y_test, y_pred)
score
```
