# Vector operations

u * v in numpy is an element-wise product, not a dot product (vTu)

```python
u = np.array([2, 4, 5, 6])
v = np.array([1, 0, 0, 2])
```

```python
u * v
```

```python
#def dot(u, v):
#    assert u.shape[0] == v.shape[0]
#    return sum(u[i] * v[i] for i in range(len(u)))

# it already exists in numpy
u.dot(v)
```

```python
# matrix vector product Uv
U = np.array([[2, 4, 5, 6],
              [1, 2, 1, 2],
              [3, 1, 2, 1]])

U.dot(v)
```

```python
# matrix matrix product UV
V= np.array([[1, 1, 2],
             [0, 0.5, 1],
             [0, 2, 1],
             [2, 1, 0]])
U.dot(V)
```

```python
U @ V
```

```python
# identity matrix
I = np.eye(4)
I
```

```python
U.dot(I)
```

```python
# inverse
Vs = V[[0, 1, 2]]
Vs
```

```python
Vs_inv = np.linalg.inv(Vs)
Vs_inv
```

```python
Vs_inv.dot(Vs)
```

```python
Vs.dot(Vs_inv)
```
